```python
y = x ** 3
plt.plot(x, y, color='purple', linewidth = 5)
plt.xlabel("x values")
plt.ylabel("Cube of x")
plt.title("Matplotlib demo")
plt.show()
```

```python
# color, marker, linewidth, labels for axis, datasets and the title
y = x ** 3
y2 = x ** 2
plt.plot(x, y, label="Cube of x", color='purple', linewidth = 5)
plt.plot(x, y2, label="Square of x", color='olive', linewidth = 5)
plt.xlabel("x values")
plt.ylabel("Y values")
plt.title("Matplotlib demo")
plt.legend()
plt.show()
```

```python
# setting up own axis
y = x ** 3
y2 = x ** 2
plt.plot(x, y, label="Cube of x", color='purple', linewidth = 5)
plt.plot(x, y2, label="Square of x", color='olive', linewidth = 5)
plt.axis([0, 10, 0, 200])
plt.xlabel("x values")
plt.ylabel("Y values")
plt.title("Matplotlib demo")
plt.legend()
plt.show()
```

```python
# setting up grid
y = x ** 3
y2 = x ** 2
plt.plot(x, y, label="Cube of x", color='purple', linewidth = 5)
plt.plot(x, y2, label="Square of x", color='olive', linewidth = 5)
plt.xlabel("x values")
plt.ylabel("Y values")
plt.title("Matplotlib demo")
plt.grid()
plt.legend()
plt.show()
```

```python
# add random text at some place in the graph
y = x ** 3
y2 = x ** 2
plt.plot(x, y, label="Cube of x", color='purple', linewidth = 5)
plt.plot(x, y2, label="Square of x", color='olive', linewidth = 5)
plt.xlabel("x values")
plt.ylabel("Y values")
plt.title("Matplotlib demo")
plt.grid()
plt.text(2, 80, "Customized graph", fontsize=14)
plt.legend()
plt.show()
```

```python
# Pie graphs
sizes = [3, 4, 6, 2]
plt.pie(sizes)
plt.show()
```

```python
# making the graph round
sizes = [3, 4, 6, 2]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
# change colors
sizes = [3, 4, 6, 2]
colors = ["purple", "orange", "blue", "pink"]
plt.pie(sizes, colors=colors)
plt.axis("equal")
plt.show()
```

```python
# Adding labels and title
labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["purple", "orange", "blue", "pink"]
plt.pie(sizes, colors=colors, labels = labels)
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

```python
# Adding text to pie sections in the pie graph
labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["purple", "orange", "blue", "pink"]
plt.pie(sizes, colors=colors, labels = labels, autopct="%.2f%%")
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

```python
# explode a section in the pie
labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["purple", "orange", "blue", "pink"]
explode = [0.1, 0, 0, 0]
plt.pie(sizes, colors=colors, labels = labels, explode=explode, autopct="%.2f%%")
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

```python
# By default pie graph is counter-clockwise
# Changing to clockwise
labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["purple", "orange", "blue", "pink"]
explode = [0.1, 0, 0, 0]
plt.pie(sizes, colors=colors, labels = labels, explode=explode, autopct="%.2f%%", counterclock=False)
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

```python
# Change from where first entry (pie) starts in the pie graph
labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["purple", "orange", "blue", "pink"]
explode = [0.1, 0, 0, 0]
plt.pie(sizes, colors=colors, labels = labels, explode=explode, autopct="%.2f%%", startangle=100)
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

```python
labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["purple", "orange", "blue", "pink"]
explode = [0.1, 0, 0, 0]
plt.pie(sizes, colors=colors, labels = labels, explode=explode, autopct="%.2f%%", counterclock=False, startangle=100)
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```
