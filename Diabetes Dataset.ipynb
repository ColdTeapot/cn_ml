#### Training an Algorithm

```python
from sklearn import model_selection
```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Verify that the split data points adds up to equal the total data points
X_train.shape[0] + X_test.shape[0] == X.shape[0]
```

```python
from sklearn.linear_model import LinearRegression
algo = LinearRegression()
```

```python
# Train the data by asking to fit
algo.fit(X_train, Y_train)
```

```python
Y_pred = algo.predict(X_test)
Y_pred
```

```python
# There are many ways of figuring how the algorithm is performing - One way is by plotting
# Compare Y_pred and Y_test by plotting
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(Y_test, Y_pred)  # Keep Y_test (true values) on X-axis and Y_pred values on Y-axis
plt.axis([0,350, 0,350])
plt.show()
```
